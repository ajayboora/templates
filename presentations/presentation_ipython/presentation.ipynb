# Code example

```python
from pylab import *

x = linspace(0, 1)

figure()
for power in [1, 2, 3, 4, 5]:
    plot(x, x**power, label='x**{0}'.format(power))
legend(loc='best')
xlabel('x')
ylabel('x**power')
title('Different powers')
show()
```

# Interactive widget example
Interactivity still doesn't work; if interactivity needed, just drag another browser window into place.

```python
%matplotlib inline
from __future__ import print_function
from pylab import *
from IPython.html.widgets import interactive
import IPython.html.widgets as widgets
from IPython.display import display
```

```python
def f(f0=1):
    t = linspace(0, 1, num=1e3)
    figure()
    plot(t, sin(2*pi*t*f0))
    show()
    
f0_slider = widgets.FloatSliderWidget(min=1, max=10, step=0.5, value=1)

w = interactive(f, f0=f0_slider)
display(w)
```
